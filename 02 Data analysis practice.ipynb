```python
# Divivde budget and gross columns by 1000000 to convert to millon $
df_IMDb['budget'] = df_IMDb['budget'].div(1000000)
df_IMDb['Gross'] = df_IMDb['Gross'].div(1000000)
df_IMDb
```

```python
# Create new column as fifth column named 'profit' by subtracting budget from gross
df_IMDb.insert(4, column = 'profit', value = df_IMDb['Gross'] - df_IMDb['budget'])
df_IMDb
```

```python
# Sort the dataframe with 'profit' column
df_IMDb.sort_values(by = ['profit'], ascending = False, inplace = True, ignore_index = True)
df_IMDb
```

```python
# Show up the name of top 6 profitable movies 
top6 = df_IMDb.loc[0:5,'Title']
top6
```

```python
df_IMDb.head(6)
```

```python
# Show up the movies with negative profit
negative_profit = df_IMDb[df_IMDb['profit'] < 0]
negative_profit.reset_index(drop = True)
```

```python
# # Show up the movies with negative profit and after 2013 year
negative_profit_2013 = df_IMDb[(df_IMDb['profit'] < 0) & (df_IMDb['title_year'] > 2013)]
negative_profit_2013
```

```python
# Plot budget and profit
sns.set_style('whitegrid')
plt.figure(figsize = (10,5))
h = sns.jointplot(x = 'budget', y = 'profit', data = df_IMDb, color = (255/255., 152/255., 150/255.))
h.set_axis_labels('budget', 'profit', fontsize = 12)
plt.subplots_adjust(top=0.9)
plt.suptitle('Scatter plot of budget and profit', fontsize = 15)
plt.show()
```

```python
# Create new column named 'likes' that sums up facebook likes of all three actors
df_IMDb.insert(11, column = 'likes', value = df_IMDb.loc[:,['actor_1_facebook_likes',
                                                            'actor_2_facebook_likes',
                                                            'actor_3_facebook_likes']].sum(axis=1))
df_IMDb
```

```python
# Top 5 likes 
top5_likes = df_IMDb.sort_values(by = 'likes', ascending = False, ignore_index = True).loc[0:4,:]
top5_likes
```

```python
top5_likes = df_IMDb.sort_values(by = 'likes', ascending = False, ignore_index = True).iloc[:5,:12]
top5_likes
```

```python
# likes histogram/density plot
plt.figure(figsize = (10,5))
likes = sns.distplot(df_IMDb['likes'], color = (247/255., 189/255., 199/255.))
likes.axes.set_title('Movie likes distribution', fontsize = 15)
likes.set_xlabel('likes', fontsize = 12)
plt.show()
```

```python
# Find all movies in 2015
def is_2015(title_year):
    if title_year == 2015:
        return True
    else: 
        return False
    
    
# Find all movies in USA
def is_USA(Country):
    if 'USA' in Country:
        return True
    else:
        return False

    
df_IMDb['title_year'].apply(is_2015)
```

```python
df_IMDb['Country'].apply(is_USA)
```
